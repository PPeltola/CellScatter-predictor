```python
import json
import matplotlib.pyplot as plt
from cellscatter_predictor import CellScatterPredictor

# Load the data:
f = open("simulation_formfactors.json", "r")
formfactors = json.load(f)

print(len(formfactors))
print(len(formfactors[0]))
```

The example data point set has 583 simulated formfactors and every formfactor is 1000 values long. A plot of a single example formfactor:

```python
# Plot an example data point
test_ff = formfactors[0]

plt.plot(test_ff)
plt.show()
```

Below are the example use cases shown in the README file as well:

```python
csp = CellScatterPredictor()
preds = csp.predict(test_ff)
```

```python
print(preds.keys())
print(preds['thickness'])
print(preds['APL'])
```

```python
print(type(preds['density']))
print(len(preds['density']))
print(preds['density'][0])
```

```python
thickness = csp.predict_thickness(test_ff, print_text=False)
apl = csp.predict_apl(test_ff, print_text=False)
density = csp.predict_density(test_ff, plot=False)

print(thickness)
print(apl)
print(len(density))
```

```python
preds = csp.predict(test_ff, plot=False, print_text=False)
print(preds['density'][30:33])
```
